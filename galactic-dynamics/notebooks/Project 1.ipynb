```python
from astropy.io import ascii
import astropy.coordinates as coord
import astropy.units as u
import emcee
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
import scipy.optimize as so
import triangle
%matplotlib inline
```

```python
hip = ascii.read("../data/hipparcos.txt")
print(hip.colnames)
```

We'll start by defining some cuts to remove (as best we can) bad or useless data from the sample.

```python
distance = (1000./hip['parallax'])*u.pc
c = coord.Galactic(l=hip['l']*u.degree, b=hip['b']*u.degree, distance=distance)

clean_idx = ((hip['parallax'] > 0.) & 
             ((hip['err_parallax'] / hip['parallax']) < 0.1) & # less than 10% parallax uncertainty
             (hip['multiplicity_flag'] == 0) & # ensure no binaries are in the sample
             (np.abs(c.cartesian.z) <= (50*u.pc)) # near the plane
             )
```

```python
clean_data = hip[clean_idx]
print("{} stars left after cuts.".format(len(clean_data)))
```

```python
# convert parallax in mas to arcseconds, convert to pc
distance = (1000./clean_data['parallax'])*u.pc
plt.hist(distance.value, bins=np.linspace(0.,250.,100))
plt.xlabel("Distance [pc]")
```

# Problem 1

We'll make an HR diagram for all of the stars left after the above data quality cuts

```python
M_V = clean_data['Vmag'] - 5*np.log10(distance.to(u.pc).value) + 5

fig,axes = plt.subplots(1,2,figsize=(10,5),sharex=True, sharey=True)
axes[0].plot(clean_data['BminusV'], M_V, linestyle='none', alpha=0.2)
axes[0].set_xlim(-0.5,2.)
axes[0].set_ylim(15,-5)
axes[0].set_xlabel("$B-V$")
axes[0].set_ylabel("$M_V$")

H,xedges,yedges = np.histogram2d(clean_data['BminusV'], M_V, bins=100)
axes[1].pcolormesh(xedges, yedges, H.T, cmap=cm.cubehelix_r)
axes[1].set_xlabel("$B-V$")

fig.tight_layout()
```

# Problem 2

(Derivation not shown)

$$
\mu_l = A\cos{2l} + B
$$

```python
cos2l = np.cos(2*clean_data['l']*np.pi/180.)
```

```python
def model(A, B, cos2l):
    return (A*cos2l + B) / 4.7405

def ln_prior(p):
    """ Parameters are the Oort constants, A, B, and the
        velocity dispersion of the stars. 
    """
    A,B,vel_disp = p
    
    lnp = 0.
    
    if A < 0.:
        return -np.inf
    
    # Make sure circular velocity isn't crazy
    V0 = 8.*(A - B)
    if V0 < 170 or V0 > 250:
        return -np.inf
    
    # Make sure velocity dispersion is positive
    if vel_disp < 0.:
        return -np.inf
    
    # Prior on a variance should be p(σ) ~ 1/σ
    lnp += -np.log(vel_disp)
    
    return lnp

def ln_likelihood(p, cos2l, mul, err_mul, distance):
    A,B,vel_disp = p
    
    # Convert velocity dispersion to extra scatter in proper motion
    scatter = (vel_disp*u.km/u.s / distance).to(u.mas/u.yr, equivalencies=u.dimensionless_angles()).value
    var = err_mul**2 + scatter**2
    lnlike = -0.5*np.log(var) - 0.5*( (mul - model(A, B, cos2l))**2 / var )
    
    return lnlike

def ln_posterior(p, *args):
    prior = ln_prior(p)
    if np.any(np.isinf(prior)):
        return -np.inf
    
    return prior + ln_likelihood(p, *args).sum()
```

```python
nwalkers = 16
p0 = emcee.utils.sample_ball([10.,-10.,21], [1.,1.,1.], size=nwalkers)
sampler = emcee.EnsembleSampler(nwalkers=nwalkers, dim=p0.shape[1], lnpostfn=ln_posterior, 
                                args=(cos2l,clean_data['mu_l'],clean_data['err_mu_l'],(1000./clean_data['parallax'].data)*u.pc))
```

```python
sampler.reset()
pos,prob,state = sampler.run_mcmc(p0, 200)
sampler.reset()
pos,prob,state = sampler.run_mcmc(pos, 1000)
```

```python
for chain in sampler.chain[...,0]:
    plt.plot(chain, drawstyle='steps', color='k', alpha=0.1)
```

```python
fig = triangle.corner(sampler.flatchain, truths=[14.82, -12.37, None], 
                      labels=["$A$", "$B$", r"$\sigma_v$"])
```

```python
best_params = np.median(sampler.flatchain, axis=0)
std_params = np.std(sampler.flatchain, axis=0)
print("Fit A: {:.2f} +/- {:.2f} km/s/kpc".format(best_params[0], std_params[0]))
print("Fit B: {:.2f} +/- {:.2f} km/s/kpc".format(best_params[1], std_params[1]))
print("Fit σ: {:.2f} +/- {:.2f} km/s".format(best_params[2], std_params[2]))
```

# Repeat with main sequence stars

We select main sequence stars with some rough lines...

```python
def top_cut(x):
    y = np.zeros_like(x)
    y[(x < 0.55)] = 6*x[(x < 0.55)] - 2
    y[(x > 0.55) & (x < 0.78)] = 10*x[(x > 0.55) & (x < 0.78)] - 3.8
    y[(x > 0.78) & (x < 1.5)] = 5*x[(x > 0.78) & (x < 1.5)] + 0.2
    y[(x > 1.5)] = 10*x[(x > 1.5)] - 6.
    return y

def bottom_cut(x):
    # y = np.zeros_like(x)
    return 6*x + 3
```

```python
M_V = clean_data['Vmag'] - 5*np.log10(distance.to(u.pc).value) + 5

fig,axes = plt.subplots(1,2,figsize=(10,5),sharex=True, sharey=True)
axes[0].plot(clean_data['BminusV'], M_V, linestyle='none', alpha=0.2)

x = np.linspace(-0.5,2,100)
axes[0].plot(x, top_cut(x), linestyle='-', marker=None, lw=2.)
axes[0].plot(x, bottom_cut(x), linestyle='-', marker=None, lw=2.)

axes[0].set_xlim(-0.5,2.)
axes[0].set_ylim(15,-5)
axes[0].set_xlabel("$B-V$")
axes[0].set_ylabel("$M_V$")

H,xedges,yedges = np.histogram2d(clean_data['BminusV'], M_V, bins=100)
axes[1].pcolormesh(xedges, yedges, H.T, cmap=cm.cubehelix_r)
axes[1].set_xlabel("$B-V$")

fig.tight_layout()
```

```python
main_seq_idx = (clean_data['Vmag'] > top_cut(clean_data['BminusV'])) & \
               (clean_data['Vmag'] < bottom_cut(clean_data['BminusV']))
```

```python
main_seq = clean_data[main_seq_idx]
print(len(main_seq))

cos2l = np.cos(2*main_seq['l']*np.pi/180.)
```

```python
nwalkers = 16
p0 = emcee.utils.sample_ball([10.,-10.,21], [1.,1.,1.], size=nwalkers)
sampler = emcee.EnsembleSampler(nwalkers=nwalkers, dim=p0.shape[1], lnpostfn=ln_posterior, 
                                args=(cos2l,main_seq['mu_l'],main_seq['err_mu_l'],(1000./main_seq['parallax'].data)*u.pc))
```

```python
pos,prob,state = sampler.run_mcmc(p0, 200)
sampler.reset()
pos,prob,state = sampler.run_mcmc(pos, 1000)
```

```python
fig = triangle.corner(sampler.flatchain, truths=[14.82, -12.37, None], 
                      labels=["$A$", "$B$", r"$\sigma_v$"])
```

```python
best_params = np.median(sampler.flatchain, axis=0)
std_params = np.median(sampler.flatchain - np.median(sampler.flatchain,axis=0), axis=0)
print("Fit A: {:.2f} +/- {:.2f} km/s/kpc".format(best_params[0], std_params[0]))
print("Fit B: {:.2f} +/- {:.2f} km/s/kpc".format(best_params[1], std_params[1]))
print("Fit σ: {:.2f} +/- {:.2f} km/s".format(best_params[2], std_params[2]))
```
