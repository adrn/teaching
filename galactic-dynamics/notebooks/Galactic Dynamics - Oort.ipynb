```python
# figure out selection in |z| by ~0.5 kpc

# convert parallax in mas to arcseconds, convert to pc
distance = ((1000./hip['parallax'])*u.pc).to(u.kpc)

c = coord.Galactic(l=hip['l']*u.degree, b=hip['b']*u.degree, distance=distance)
```

```python
ix = ((hip['err_parallax'] / hip['parallax']) < 0.1) & \
     (distance > 100*u.pc) & (distance < 1000*u.pc) & \
     (np.abs(c.cartesian.z) <= 0.2*u.kpc) & \
     ((hip['err_mu_l'] / hip['mu_l']) < 0.25)
len(hip), sum(ix)
```

```python
subsample = hip[ix]

plt.plot(np.cos(2*subsample['l']*np.pi/180.), subsample['mu_l'], linestyle='none')
plt.xlim(-1.,1.)
```

```python
# sigma clipping
for ii in range(100):
    sigma = np.median(np.abs(subsample['mu_l'] - np.median(subsample['mu_l'])))
    clip_ix = (subsample['mu_l'] < (np.median(subsample['mu_l']) + 5*sigma)) & \
              (subsample['mu_l'] > (np.median(subsample['mu_l']) - 5*sigma))
    subsample = subsample[clip_ix]
    
    if sum(clip_ix) == 0:
        break
```

```python
cos2l = np.cos(2*subsample['l']*np.pi/180.)
```

```python
# the "y" data are the proper motions, the "x" data are the galactic longitudes
def model(A, B, cos2l):
    return (A*cos2l + B) / 4.7405

def ln_prior(p):
    A,B,vel_scatter,Poutlier = p
    
    lnp = 0.
    
    if A < 0.:
        return -np.inf
    
    V0 = 8.*(A - B)
    if V0 < 170 or V0 > 240:
        return -np.inf
    
    if vel_scatter < 0.:
        return -np.inf
    
    lnp += -np.log(vel_scatter)
    
    if Poutlier < 0. or Poutlier > 1.:
        return -np.inf
    
    return lnp

def ln_likelihood(p, cos2l, mul, err_mul, distance):
    A,B,vel_scatter,Poutlier = p
    scatter = (vel_scatter*u.km/u.s / distance).to(u.mas/u.yr, equivalencies=u.dimensionless_angles()).value
    lnlike1 = -0.5*np.log(err_mul**2 + scatter**2) - 0.5*( (mul - model(A, B, cos2l))**2 / (err_mul**2 + scatter**2) )
    
    outlier_sigma = 10. # TODO: maybe this is too large...
    lnlike2 = -np.log(outlier_sigma) - 0.5*(mul / outlier_sigma)**2
        
    lnlike = np.vstack((lnlike1,lnlike2))
    
    return logsumexp(lnlike, axis=0, b=np.array([[1-Poutlier], [Poutlier]]))

def ln_posterior(p, *args):
    prior = ln_prior(p)
    if np.any(np.isinf(prior)):
        return -np.inf
    
    return prior + np.sum(ln_likelihood(p, *args))
```

```python
import emcee, triangle
```

```python
nwalkers = 16
p0 = emcee.utils.sample_ball([10.,-10.,21,0.5], [1.,1.,1.,0.1], size=nwalkers)
sampler = emcee.EnsembleSampler(nwalkers=nwalkers, dim=p0.shape[1], lnpostfn=ln_posterior, 
                                args=(cos2l,subsample['mu_l'],subsample['err_mu_l'],(1000./subsample['parallax'].data)*u.pc))
```

```python
pos,prob,state = sampler.run_mcmc(p0, 200)
sampler.reset()
pos,prob,state = sampler.run_mcmc(pos, 1000)
```

```python
fig = triangle.corner(sampler.flatchain, truths=[14.82, -12.37, None, None])
```
