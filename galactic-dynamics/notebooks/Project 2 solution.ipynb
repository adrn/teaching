# Problem 1

```python
def leapfrog_integrate(acceleration_func, x0, v0, dt, nsteps, args=()):
    """ A simple implementation of Leapfrog integration 
    
    Parameters
    ----------
    acceleration_func : callable
        A function that computes the acceleration at a given position.
    x0 : numpy.ndarray
        Initial position(s).
    v0 : numpy.ndarray
        Initial velocity(s).
    dt : numeric
        Timestep.
    nsteps : int
        Number of steps to run for.
    args : iterable (optional)
        Any other arguments the acceleration function takes.
        
    Returns
    -------
    t : numpy.ndarray
        Array of times.
    x : numpy.ndarray
        Array of positions.
    v : numpy.ndarray
        Array of velocities.
    """
    
    # ensure that the initial conditions are arrays and at least 2D
    x0 = np.atleast_2d(x0)
    v0 = np.atleast_2d(v0)
    norbits,ndim = x0.shape
    
    # wrapper around the acceleration function so we can call it with just the position, x
    acc = lambda x: acceleration_func(x, *args)
    
    all_x = np.zeros((nsteps,norbits,ndim))
    all_v = np.zeros((nsteps,norbits,ndim))
    t = np.zeros(nsteps)
    
    all_x[0] = x0
    all_v[0] = v0
    
    # velocity at 1/2 step 
    v_iminus1_2 = v0 + acc(x0)*dt/2.
    x_iminus1 = x0.copy()
    for i in range(1,nsteps):
        x_i = x_iminus1 + v_iminus1_2*dt # full step
        a_i = acc(x_i)
        v_i = v_iminus1_2 + a_i*dt/2. # half step
        v_iplus1_2 = v_i + a_i*dt/2. # half step
        
        all_x[i] = x_i
        all_v[i] = v_i
        t[i] = t[i-1] + dt
        
        x_iminus1 = x_i
        v_iminus1_2 = v_iplus1_2
    
    return t, all_x, all_v
```

Here we define a function to compute the acceleration due to a simple harmonic oscillator potential,
$$
\Phi(x) = \frac{1}{2}\omega^2 x^2\\
a_x = -\frac{\partial \Phi}{\partial x} = -\omega^2 x
$$

```python
def sho_acceleration(x, omega):
    return -omega**2 * x
```

```python
t,x,v = leapfrog_integrate(sho_acceleration, x0=[1.], v0=[0.], args=(2.,), dt=0.1, nsteps=10000)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x[:,0,0], v[:,0,0], marker=None)
plt.xlim(-2.1,2.1)
plt.ylim(-2.1,2.1)
plt.xlabel("$x$")
plt.ylabel("$v_x$")
```

# Problem 2

I'll use a potential class that I've already defined for a Miyamoto-Nagai disk. We can get the circular velocity at a given position as:
$$
v_c(r_0) = \sqrt{r_0 \frac{d \Phi}{d r}(r_0)}
$$
or, in the midplane,
$$
v_c(x_0) = \sqrt{x_0 \frac{d \Phi}{d x}(x_0)}
$$

```python
import gary.potential as gp
from gary.units import galactic
from scipy.optimize import fmin
```

```python
Rsun = 8.3 # kpc
vcirc_sun = 220. # km/s
```

```python
def fit_func(m):
    disk = gp.MiyamotoNagaiPotential(m=m, a=6.5, b=0.26, units=galactic)
    dPhi_dx = disk.gradient([[Rsun,0,0]])[0,0]
    vcirc = np.sqrt(np.abs(Rsun * dPhi_dx))
    vcirc = (vcirc*u.kpc/u.Myr).to(u.km/u.s).value
    return np.abs(vcirc - vcirc_sun)
```

```python
m = fmin(fit_func, x0=2.5E11)[0]
```

```python
disk = gp.MiyamotoNagaiPotential(m=m, a=6.5, b=0.26, units=galactic)
```

```python
# plot a rotation curve
xs = np.linspace(1., 20., 100)
r = np.zeros((len(xs),3))
r[:,0] = xs

dPhi_dx = disk.gradient(r)[:,0]
vcirc = np.sqrt(np.abs(Rsun * dPhi_dx))

plt.plot(xs, (vcirc*u.kpc/u.Myr).to(u.km/u.s), marker=None)
plt.axvline(Rsun)
plt.xlabel(r"$x$ ${\rm [kpc]}$")
plt.ylabel(r"$v_{\rm circ}$ ${\rm [km \, s^{-1}]}$")
```

```python
from scipy.signal import argrelmax
```

```python
w0 = np.vstack((x.to(u.kpc), v.to(u.kpc/u.Myr))).T.copy()
t,w = disk.integrate_orbit(w0.copy(), dt=0.5, nsteps=4000)
```

```python
plt.figure(figsize=(8,5))
for i in range(3):
    R = np.sqrt(np.sum(w[:,i,:2]**2, axis=-1))
    z = w[:,i,2]
    
    plt.plot(R, z, marker='.', alpha=0.25, linestyle='none')

plt.xlim(5,10)
plt.ylim(-1,1)
plt.xlabel("$R$ [kpc]")
plt.ylabel("$z$ [kpc]")
```

```python
orbit_r = np.sqrt(np.sum(w[...,:3]**2, axis=-1)) # spherical radius / distance
orbit_R = np.sqrt(np.sum(w[...,:2]**2, axis=-1)) # cylindrical radius
orbit_phi = np.arctan2(w[...,1],w[...,0]) # azimuthal angle
orbit_z = w[...,2] # z / height above the midplane
```

```python
pericenters = orbit_r.min(axis=0)
apocenters = orbit_r.max(axis=0)
max_height = np.abs(orbit_z).max(axis=0)
```

```python
n = orbit_r.shape[1]
T_R = np.zeros(n)
T_phi = np.zeros(n)
T_z = np.zeros(n)
failed_ix = []
for i in range(n):
    try:
        # radial period
        derp = argrelmax(orbit_R[:,i])[0]
        # (t[derp[1:]] - t[derp[:-1]]).mean()
        T_R[i] = t[derp[1]] - t[derp[0]] 

        # azimuthal period
        derp = argrelmax(orbit_phi[:,i])[0]
        T_phi[i] = t[derp[1]] - t[derp[0]]

        # vertical period
        derp = argrelmax(orbit_z[:,i])[0]
        T_z[i] = t[derp[1]] - t[derp[0]]
    except:
        print("Failed to compute periods for orbit: {}".format(i))
        failed_ix.append(i)
        continue
        
failed_ix = np.array(failed_ix)
```

```python
pericenters = np.delete(pericenters, failed_ix)
apocenters = np.delete(apocenters, failed_ix)
max_height = np.delete(max_height, failed_ix)
T_R = np.delete(T_R, failed_ix)
T_phi = np.delete(T_phi, failed_ix)
T_z = np.delete(T_z, failed_ix)
```

```python
fig,axes = plt.subplots(2,3,figsize=(16,12))

bins = np.linspace(0.,20,50)
axes[0,0].hist(pericenters, bins=bins)
axes[0,0].set_xlabel(r"$r_{\rm peri}$ [kpc]")
axes[0,1].hist(apocenters, bins=bins)
axes[0,1].set_xlabel(r"$r_{\rm apo}$ [kpc]")

bins = np.linspace(0.,1000,25)
axes[0,2].hist(max_height*1000., bins=bins)
axes[0,2].set_xlabel(r"$\max(|z|)$ [pc]")

# -- periods --
axes[1,0].hist(T_R, bins=np.linspace(50.,250,25))
axes[1,0].set_xlabel(r"$T_{R}$ [Myr]")

axes[1,1].hist(T_phi, bins=np.linspace(125.,325,25))
axes[1,1].set_xlabel(r"$T_{\phi}$ [Myr]")

axes[1,2].hist(T_z, bins=np.linspace(0.,100,25))
axes[1,2].set_xlabel(r"$T_{z}$ [Myr]")
```
