```python
import streamteam.dynamics as sd
import streamteam.potential as sp
import streamteam.integrate as si
from streamteam.units import galactic
import scipy.signal as ss
import scipy.integrate as sci
```

```python
# potential = sp.LogarithmicPotential(v_c=VCIRC.decompose(usys).value, r_h=0., 
#                                     q1=1, q2=1, q3=1, phi=0.)
# potential = sp.MiyamotoNagaiPotential(m=2.1E11, a=6.5, b=0.26, units=galactic)
potential = sp.IsochronePotential(m=1.2E12, b=10., units=galactic)
(np.sqrt(8*potential.gradient(np.array([[8.,0,0]])))*u.kpc/u.Myr).to(u.km/u.s)
```

```python
ts,ws = potential.integrate_orbit(w, dt=0.5, nsteps=1000, Integrator=si.DOPRI853Integrator)
```

```python
xp = x.copy()
vp = v.copy()
xp[:,2] = 0.
vp[:,2] = 0.
E = potential.energy(xp, vp)
L = sd.angular_momentum(w)
Lz = L[:,2]
Lmag = np.linalg.norm(L, axis=1)
```

Measure radial and vertical periods

```python
Tr = np.zeros_like(E)
Tp = np.zeros_like(E)
Tz = np.zeros_like(E)
for i in range(w.shape[0]):
    apo_idxs, = ss.argrelmax(np.linalg.norm(ws[:,i,:3], axis=1), axis=0)
    dt = ts[apo_idxs[1:]] - ts[apo_idxs[:-1]]
    Tr[i] = np.median(dt)
    
    r_a = ws[apo_idxs[0],i,:3]
    r_b = ws[apo_idxs[1],i,:3]
    dphi = (np.arccos(r_a.dot(r_b) / np.linalg.norm(r_a) / np.linalg.norm(r_b))*u.rad).to(u.degree).value
    Tp[i] = 2*np.pi/np.abs(dphi)*Tr[i]
    
    z_idxs, = ss.argrelmax(ws[:,i,2], axis=0)
    dt = ts[z_idxs[1:]] - ts[z_idxs[:-1]]
    Tz[i] = np.median(dt)
T = np.vstack((Tr,Tp,Tz)).T
Omega = 2*np.pi/T
```

Approximate motion in Z direction as uncoupled:

```python
Ez = 0.5*(v[:,2]**2 + x[:,2]**2*Omega[:,2]**2)
```

```python
r = np.linalg.norm(ws[...,:3], axis=-1)
pers = r.min(axis=0)
apos = r.max(axis=0)
```

```python
def rdot(r, E, Lz):
    x = [r,0,0]
    return np.sqrt(2*(E - potential.value(x)) - Lz**2/r**2)

J = np.zeros_like(Omega)
for i in range(w.shape[0]):
    sol = sci.quad(rdot, pers[i], apos[i]-0.01, args=(E[i], Lz[i]))
    J[i,0] = sol[0] / 2*np.pi

J[:,1] = Lz
J[:,2] = Ez / Omega[:,2]
```
