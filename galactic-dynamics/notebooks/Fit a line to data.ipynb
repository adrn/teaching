```python
# imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from scipy import optimize
import emcee
import triangle
from scipy.stats import multivariate_normal
```

# Fitting a line to data

This is not meant to be a comprehensive review of the statistics involved with fitting a model (straight line) to a set of data. For an excellent review on this topic, see [this paper](http://arxiv.org/abs/1008.4686).

---

We'll start by generating some fake data points $(x_i, y_i)$. We're going to assume that we have _perfect knowledge_ of the $x_i$ values but measure the $y_i$ with some _heteroscedastic_ uncertainties, $\sigma_i$. This is analogous to, say, fitting the light curve of a faint star where we know the measurement times very precisely but the fluxes are not well measured. We further assume that without uncertainties, the data, $y_i$, would have be drawn from a _perfectly thin_ straight line with parameters $(m,b)$, and the only reason the values deviate from the line is that there was some offset due to noise drawn from a normal (Gaussian) distribution of mean, $\mu=0$ and variances given by the $\sigma_i^2$. Lastly, we assume that the data are independent.

You may use `ndata` to control the number of data points and `pars` to set the true parameters of the model, or just stick with the defaults I set.

```python
ndata = 25
true_pars = (4.5, 1.7)

np.random.seed(42)
x = np.random.uniform(0., 10., size=ndata)
y = true_pars[0]*x + true_pars[1]

sigma_y = np.random.uniform(0.1, 0.2, size=ndata)*y.max() # uncertainties
y += np.random.normal(scale=sigma_y)
```

```python
# plot the generated data
plt.figure(figsize=(8,6))
plt.errorbar(x, y, sigma_y, marker='o', linestyle='none', ecolor='#bbbbbb', 
             elinewidth=2, markersize=8)
plt.xlabel("x")
plt.ylabel("y")
```

### Using least-squares

With all of the the above assumptions, the procedure of finding "best-fit" values for $m$ and $b$ is the standard process of minimizing the total squared error, scaled by the uncertainties, $\chi^2$:

$$
\chi^2 = \sum_i \frac{(y_i-f(x_i))^2}{\sigma_i^2}
$$
where
$$
f(x_i) = mx_i + b
$$

We could solve this analytically with linear algebra (and it is an _excellent_ exercise), but here I will simply use the `leastsq` optimizer built in to `scipy`. This function, `scipy.optimize.leastsq`, takes a function as its first argument that computes $\sqrt{\chi^2}$ for each data point and returns a vector of values (then it internally squares and sums). You must also pass in an initial parameter guess (`x0`), and any other information that the model needs (`args`), e.g. the observed $y_i$, $x_i$, and $\sigma_i$.

```python
def model(pars, x):
    m,b = pars
    return m*x + b

def error_function(pars, x, y, sigma_y):
    return (y - model(pars, x)) / sigma_y
```

```python
initial_parameter_guess = [1., 0.]
output = optimize.leastsq(error_function, x0=initial_parameter_guess,
                          args=(x, y, sigma_y), full_output=True)
```

With the `full_output` argument set to true, `leastsq()` returns a bunch of stuff. Of interest to us are the first two things returned: the best fit parameters and the covariance matrix of the parameters which specifies the uncertainties and correlations over the derived parameters.

```python
best_pars = output[0]
cov = output[1]

# create a distribution object for the parameters
dist = multivariate_normal(mean=best_pars, cov=cov) 
```

```python
# plot the generated data and best-fit line
plt.figure(figsize=(8,6))
plt.errorbar(x, y, sigma_y, marker='o', linestyle='none', ecolor='#bbbbbb', 
             elinewidth=2, markersize=8)

x_l = np.linspace(0., 10., 100)

# draw 100 samples from the inferred parameter distribution and plot over the data
param_samples = dist.rvs(size=100)
for params in param_samples:
    plt.plot(x_l, model(params, x_l), marker=None, 
             linestyle='-', color='#2166AC', alpha=0.2)
    
plt.plot(x_l, model(best_pars, x_l), marker=None, label='best fit',
         linestyle='-', color='#B2182B', linewidth=3.)
plt.plot(x_l, model(true_pars, x_l), marker=None, label='true',
         linestyle='-', color='#006837', linewidth=3.)
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc='upper left')
```

In the plot above, the blue lines represent individual samples from the inferred parameter distribution. The red line is the best-fit value of the parameters. The green line is the true line that generated the data. Try going back up to the top, increasing the number of data points, and see what happens to the best-fit line relative to the true line.

### Using `emcee`

To use `emcee` or any MCMC method to infer the posterior probability distribution over the parameters, we must first write down a _likelihood function_, which is the probability of the data given the model. For a single data point, this is simply a Gaussian:

$$
\begin{align}
    \mathcal{L}_i &= p(y_i \, | \, m, b, x_i, \sigma_i)\\
    &= \frac{1}{\sigma_i\sqrt{2\pi}} \exp{\left[ -\frac{(y_i - f(x_i))^2}{2\sigma_i^2} \right]}
\end{align}
$$

Because we are assuming the data are independent, we can write the joint likelihood over all data as the product of the individual likelihoods:
$$
\mathcal{L} = \prod_i^N \mathcal{L}_i
$$

For some choices of parameters, this likelihood will be extremely small (because of the exponential) and can cause numerical problems on computers. It is therefore often easier to work with the log-likelihood instead:

$$
\begin{align}
    \ln \mathcal{L} &= \sum_i^N \ln \mathcal{L}_i\\
    \ln \mathcal{L} &= \sum_i^N \ln \left( \frac{1}{\sigma_i\sqrt{2\pi}} \exp{\left[ -\frac{(y_i - f(x_i))^2}{2\sigma_i^2} \right]}\right)\\
    &= -\frac{1}{2}\sum_i^N \left(\ln{(2\pi \sigma_i^2)} + \frac{(y_i - f(x_i))^2}{\sigma_i^2} \right)\\
    &= -\frac{1}{2}\sum_i^N \ln{(2\pi \sigma_i^2)} - \frac{1}{2}\sum_i^N \frac{(y_i - f(x_i))^2}{\sigma_i^2} \\
    &= K - \frac{1}{2}\sum_i^N \frac{(y_i - f(x_i))^2}{\sigma_i^2}
\end{align}
$$

In the last two lines, the sum over the uncertainties is constant for any choice of parameters, so I have replaced this sum with the constant, $K$. This should look familiar -- the log-likelihood is simply some constant minus one-half of $\chi^2$! 

For doing Bayesian inference (e.g., with MCMC), it is not enough to just specify the likelihood function (or $\chi^2$) -- we must also specify _prior_ probability functions over the parameters of the model. These functions are meant to encapsulate all prior information we have about the parameters that go in to the model. For this extremely simple example, we are going to assume uniform priors over the parameters. 

__Note:__ I am leaving out a lot of important discussion and philosophy here, so if anything is unclear please see the paper referenced at the top of this notebook, or I'd be happy to recommend many, many other references on this topic.

```python
def ln_likelihood(pars, x, y, sigma_y):
    sum_term = 2*np.log(sigma_y) + ((y - model(pars, x))/sigma_y)**2
    return -0.5*(len(x)*np.log(2*np.pi) + np.sum(sum_term))

def ln_prior(pars, x, y, sigma_y):
    m,b = pars
    
    p = 0.
    
    m_range = (1.,10.)
    if m < m_range[0] or m > m_range[1]:
        return -np.inf
    else:
        p += -np.log(m_range[1] - m_range[0])
        
    b_range = (-10.,10.)
    if b < b_range[0] or b > b_range[1]:
        return -np.inf
    else:
        p += -np.log(b_range[1] - b_range[0])
    
    return 0.

def ln_posterior(pars, *args):
    return ln_prior(pars, *args)+ ln_likelihood(pars, *args)
```

```python
nwalkers = 32
nsteps = 2000

p0 = np.zeros((nwalkers,2))
p0[:,0] = np.random.uniform(0,10,nwalkers)
p0[:,1] = np.random.uniform(0,10,nwalkers)

sampler = emcee.EnsembleSampler(nwalkers, dim=2, lnpostfn=ln_posterior, 
                                args=(x,y,sigma_y), a=1.25)
pos,prob,state = sampler.run_mcmc(p0,nsteps//10)
sampler.reset()
pos,prob,state = sampler.run_mcmc(pos,nsteps)
```

```python
titles = ['m', 'b']
for i in range(2):
    plt.figure()
    for walker in sampler.chain[...,i]:
        plt.plot(walker, drawstyle='steps', marker=None, linestyle='-', alpha=0.25)
    plt.xlabel("Step number")
    plt.ylabel(titles[i], rotation='horizontal', labelpad=12)
```

```python
fig = triangle.corner(sampler.flatchain, truths=pars, labels=['m', 'b'], 
                      extents=[(0.5,7.), (-10,10)])
```

```python
# plot the generated data and best-fit line
plt.figure(figsize=(8,6))
plt.errorbar(x, y, sigma_y, marker='o', linestyle='none', ecolor='#bbbbbb', 
             elinewidth=2, markersize=8)

x_l = np.linspace(0., 10., 100)

# draw 100 samples from the inferred parameter distribution and plot over the data
for i in np.random.randint(nsteps*nwalkers,size=100):
    plt.plot(x_l, model(sampler.flatchain[i], x_l), marker=None, 
             linestyle='-', color='#2166AC', alpha=0.2)
    
plt.plot(x_l, model(true_pars, x_l), marker=None, label='true',
         linestyle='-', color='#006837', linewidth=3.)
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc='upper left')
```
