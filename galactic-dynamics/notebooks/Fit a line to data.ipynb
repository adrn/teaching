# Fitting a line to data

This is not meant to be a comprehensive review of the statistics involved with fitting a model (straight line) to a set of data. For an excellent review on this topic, see [this paper](http://arxiv.org/abs/1008.4686).

---

```python
# imports
import numpy as np
import matplotlib.pyplot as pl
%matplotlib inline

from scipy import optimize
import emcee
import triangle
from scipy.stats import multivariate_normal
```

We first generate some fake data. Use `ndata` to control the number of data points, `pars` to set the true parameters of the model.

```python
ndata = 10
pars = (4.5, 1.7)

x = np.random.uniform(0., 10., size=ndata)
y = pars[0]*x + pars[0]

sigma_y = np.random.uniform(0.1, 0.2, size=ndata)*y.max() # uncertainties
y += np.random.normal(scale=sigma_y)
```

```python
# plot the generated data
pl.figure(figsize=(6,4))
pl.errorbar(x, y, sigma_y, marker='o', linestyle='none')
```

### Using least-squares

We first..

```python
def model(pars, x):
    m,b = pars
    return m*x + b

def error_function(pars, x, y, sigma_y):
    return (y - model(pars, x)) / sigma_y
```

```python
output = optimize.leastsq(error_function, [1., 0.], args=(x, y, sigma_y), full_output=True)
```

```python
best_pars = output[0]
cov = output[1]
dist = multivariate_normal(mean=best_pars, cov=cov)
```

```python
# plot the generated data and best-fit line
pl.figure(figsize=(6,4))
pl.errorbar(x, y, sigma_y, marker='o', linestyle='none')

x_l = np.linspace(x.min(), x.max(), 100)
pl.plot(x_l, model(best_pars, x_l), marker=None, linestyle='-')
```

### Using `emcee`

$$
\begin{align}
    \mathcal{L} &= \prod_i^N \frac{1}{\sqrt{2\pi \sigma_i^2}} \exp{\left[ -\frac{(y_i - f(x_i))^2}{2\sigma_i^2} \right]}\\
    \ln \mathcal{L} &= \sum_i^N \ln \left( \frac{1}{\sqrt{2\pi \sigma_i^2}} \exp{\left[ -\frac{(y_i - f(x_i))^2}{2\sigma_i^2} \right]}\right)\\
    &= -\frac{1}{2}\sum_i^N \left(\ln{(2\pi \sigma_i^2)} + \frac{(y_i - f(x_i))^2}{\sigma_i^2} \right)\\
    &= - \frac{1}{2} \left[ N\ln{2\pi} + \sum_i^N \left(2\ln{\sigma_i} + \frac{(y_i - f(x_i))^2}{\sigma_i^2} \right) \right]\\
\end{align}
$$

```python
def ln_likelihood(pars, x, y, sigma_y):
    sum_term = 2*np.log(sigma_y) + ((y - model(pars, x))/sigma_y)**2
    return -0.5*(len(x)*np.log(2*np.pi) + np.sum(sum_term))

def ln_prior(pars, x, y, sigma_y):
    return 0.

def ln_posterior(pars, *args):
    return ln_prior(pars, *args)+ ln_likelihood(pars, *args)
```

```python
nwalkers = 32
nsteps = 2000

p0 = np.zeros((nwalkers,2))
p0[:,0] = np.random.uniform(0,10,nwalkers)
p0[:,1] = np.random.uniform(0,10,nwalkers)

sampler = emcee.EnsembleSampler(nwalkers, dim=2, lnpostfn=ln_posterior, 
                                args=(x,y,sigma_y), a=1.25)
pos,prob,state = sampler.run_mcmc(p0,nsteps//10)
sampler.reset()
pos,prob,state = sampler.run_mcmc(pos,nsteps)
```

```python
leastsq_samples = dist.rvs(size=nwalkers*nsteps)

fig = triangle.corner(sampler.flatchain, truths=pars, labels=['m', 'b'])
fig = triangle.corner(leastsq_samples, fig=fig, 
                      contour_kwargs={'colors': u'b', 'bins': 50}, 
                      point_kwargs={'rasterized': True, 'color': u'b'},
                      hist_kwargs={'color': u'b', 'bins': 50})
```
