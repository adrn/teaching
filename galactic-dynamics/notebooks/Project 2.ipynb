# Problem 3

```python
hip = ascii.read("../data/hipparcos_rv.txt")
print(hip.colnames)
```

Quality cuts

```python
clean_idx = ((~np.isnan(hip['rv'])) & # has a measured RV
             (hip['parallax'] > 0.) & # has a measured parallax
             ((hip['err_parallax'] / hip['parallax']) < 0.05) & # less than 10% parallax uncertainty
             (hip['multiplicity_flag'] == 0) & # ensure no binaries are in the sample
             ((hip['err_mu_l'] / hip['mu_l']) < 0.05) & ((hip['err_mu_b'] / hip['mu_b']) < 0.1) # good proper motions
             )
```

```python
clean_data = hip[clean_idx]
print("{} stars left after cuts.".format(len(clean_data)))
```

```python
# convert parallax in mas to arcseconds, convert to pc
distance = (1000./clean_data['parallax'])*u.pc
plt.hist(distance.value, bins=np.linspace(0.,250.,100))
plt.xlabel("Distance [pc]")
```

```python
g = coord.Galactic(l=clean_data['l']*u.degree, b=clean_data['b']*u.degree, distance=distance)
x,v = sc.hel_to_gal_xyz(g, pm=(clean_data['mu_l']*u.mas/u.yr, clean_data['mu_b']*u.mas/u.yr), vr=clean_data['rv']*u.km/u.s,
                        xsun=-8.3*u.kpc, vcirc=220*u.km/u.s, vlsr=[11.1, 12.24, 7.25]*u.km/u.s)
```

```python
fig,axes = plt.subplots(1,2,figsize=(10,5))
axes[0].plot(x[0].to(u.kpc), x[1].to(u.kpc), marker='.', linestyle='none', alpha=0.25)
axes[0].set_xlim(-8.4,-8.2)
axes[0].set_ylim(-0.1,0.1)
axes[0].set_xlabel("$x$ [kpc]")
axes[0].set_ylabel("$y$ [kpc]")

axes[1].plot(x[0].to(u.kpc), x[2].to(u.kpc), marker='.', linestyle='none', alpha=0.25)
axes[1].set_xlim(-8.4,-8.2)
axes[1].set_ylim(-0.1,0.1)
axes[1].set_xlabel("$x$ [kpc]")
axes[1].set_ylabel("$z$ [kpc]")

fig.tight_layout()
```

```python
fig,axes = plt.subplots(1,2,figsize=(10,5))
axes[0].plot(v[0], v[1], marker='.', linestyle='none', alpha=0.25)
axes[0].set_xlim(-150,150)
axes[0].set_ylim(70, 370)
axes[0].set_xlabel(r"$v_x$ [${\rm km\,s}^{-1}$]")
axes[0].set_ylabel(r"$v_y$ [${\rm km\,s}^{-1}$]")

axes[1].plot(v[0], v[2], marker='.', linestyle='none', alpha=0.25)
axes[1].set_xlim(-150,150)
axes[1].set_ylim(-150,150)
axes[1].set_xlabel(r"$v_x$ [${\rm km\,s}^{-1}$]")
axes[1].set_ylabel(r"$v_z$ [${\rm km\,s}^{-1}$]")

fig.tight_layout()
```
