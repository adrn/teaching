```python
from astropy.constants import G
import astropy.units as u
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def leapfrog_integrate(acceleration_func, x0, v0, dt, nsteps, t1=0., args=()):
    """ A simple implementation of Leapfrog integration 
    
    Parameters
    ----------
    acceleration_func : callable
        A function that computes the acceleration at a given position.
    x0 : numpy.ndarray
        Initial position(s).
    v0 : numpy.ndarray
        Initial velocity(s).
    dt : numeric
        Timestep.
    nsteps : int
        Number of steps to run for.
    t1 : numeric (optional)
        Initial time.
    args : iterable (optional)
        Any other arguments the acceleration function takes.
        
    Returns
    -------
    t : numpy.ndarray
        Array of times.
    x : numpy.ndarray
        Array of positions.
    v : numpy.ndarray
        Array of velocities.
    """
    
    # ensure that the initial conditions are arrays and at least 2D
    x0 = np.atleast_2d(x0)
    v0 = np.atleast_2d(v0)
    norbits,ndim = x0.shape
    
    # wrapper around the acceleration function so we can call it with just the position, x
    acc = lambda t,x: acceleration_func(t, x, *args)
    
    all_x = np.zeros((nsteps,norbits,ndim))
    all_v = np.zeros((nsteps,norbits,ndim))
    t = np.zeros(nsteps)
    t[0] = t1
    
    all_x[0] = x0
    all_v[0] = v0
    
    # velocity at 1/2 step 
    v_iminus1_2 = v0 + acc(t1,x0)*dt/2.
    x_iminus1 = x0.copy()
    for i in range(1,nsteps):
        t[i] = t[i-1] + dt
        x_i = x_iminus1 + v_iminus1_2*dt # full step
        a_i = acc(t[i], x_i)
        v_i = v_iminus1_2 + a_i*dt/2. # half step
        v_iplus1_2 = v_i + a_i*dt/2. # half step
        
        all_x[i] = x_i
        all_v[i] = v_i
        
        x_iminus1 = x_i
        v_iminus1_2 = v_iplus1_2
    
    return t, all_x, all_v
```

```python
def periodic_forcing_acceleration(t, x, A, omega):
    return -np.sin(x) + A*np.cos(omega*t)
```

```python
# no forcing, regular orbit
t,x,v = leapfrog_integrate(periodic_forcing_acceleration, x0=[3.], v0=[0.], dt=0.1, nsteps=10000, args=(0., 0.75))

plt.figure(figsize=(5,5))
plt.plot(x[:,0,0], v[:,0,0], marker='.', linestyle='none', alpha=0.2)
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.xlabel(r"$\theta$")
plt.ylabel(r"$p_\theta$")
```

```python
# periodic forcing, regular orbit
t,x,v = leapfrog_integrate(periodic_forcing_acceleration, x0=[3.], v0=[0.], dt=0.1, nsteps=10000, args=(0.03, 0.15))

plt.figure(figsize=(5,5))
plt.plot((x[:,0,0] % (2*np.pi)) - np.pi, v[:,0,0], marker='.', linestyle='none', alpha=0.2)
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.xlabel(r"$\theta$")
plt.ylabel(r"$p_\theta$")
```

```python
# periodic forcing, chaotic orbit
t,x,v = leapfrog_integrate(periodic_forcing_acceleration, x0=[3.], v0=[0.], dt=0.1, nsteps=10000, args=(0.07, 0.75))

plt.figure(figsize=(5,5))
plt.plot((x[:,0,0] % (2*np.pi)) - np.pi, v[:,0,0], marker='.', linestyle='none', alpha=0.2)
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.xlabel(r"$\theta$")
plt.ylabel(r"$p_\theta$")
```
